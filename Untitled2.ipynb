```python
# Parameters for our curation
target_chembl_id = 'CHEMBL1871'
measurement_type = 'IC50'

print(f"Filtering ChEMBL36 for {target_chembl_id} {measurement_type} values...")

# Step 1: Drop old tables if they exist
%sql DROP TABLE IF EXISTS ar_raw_assays CASCADE;
%sql DROP TABLE IF EXISTS ar_curated_data CASCADE;

# Step 2: Identify High-Confidence, Non-Mutant Assays
%sql \
    SELECT a.assay_id, a.chembl_id as assay_chembl_id, a.description, a.tid, d.year \
    INTO TEMPORARY TABLE ar_raw_assays \
    FROM assays a \
    JOIN target_dictionary t ON a.tid = t.tid \
    JOIN docs d ON a.doc_id = d.doc_id \
    WHERE t.chembl_id = :target_chembl_id \
    AND a.confidence_score = 9 \
    AND a.assay_type = 'B' \
    AND (a.variant_id IS NULL) \
    AND NOT (LOWER(a.description) LIKE '%mutant%' OR LOWER(a.description) LIKE '%mutation%');

# Step 3: Extract Activities and Clean Units
%sql \
    SELECT act.molregno, act.pchembl_value, act.standard_value, act.standard_units, \
           tmp.assay_chembl_id, tmp.year as publication_year \
    INTO TEMPORARY TABLE ar_curated_data \
    FROM activities act \
    JOIN ar_raw_assays tmp ON act.assay_id = tmp.assay_id \
    WHERE act.standard_type = :measurement_type \
    AND act.standard_units = 'nM' \
    AND act.standard_relation = '=' \
    AND act.pchembl_value IS NOT NULL \
    AND act.data_validity_comment IS NULL;

# Step 4: Check our results
result = %sql SELECT COUNT(*) as total_rows FROM ar_curated_data;
print(f"Successfully curated {result} high-quality data points.")
```

```python
# 1. Enable parameter passing
%config SqlMagic.named_parameters="enabled"

# 2. Define variables
target_chembl_id = 'CHEMBL1871'
measurement_type = 'IC50'

print(f"Curating {target_chembl_id} and removing 3.0 pChEMBL unit discrepancies...")

# 3. Step 1: Filter for Single Protein Assays
%sql DROP TABLE IF EXISTS ar_raw_assays CASCADE;
%sql \
    SELECT a.assay_id, a.chembl_id as assay_chembl_id, a.description, a.tid, d.year \
    INTO TEMPORARY TABLE ar_raw_assays \
    FROM assays a \
    JOIN target_dictionary t ON a.tid = t.tid \
    JOIN docs d ON a.doc_id = d.doc_id \
    WHERE t.chembl_id = :target_chembl_id \
    AND t.target_type = 'SINGLE PROTEIN' \
    AND a.bao_format = 'BAO_0000357' \
    AND a.confidence_score = 9 \
    AND a.assay_type = 'B' \
    AND (a.variant_id IS NULL) \
    AND NOT (LOWER(a.description) LIKE '%mutant%' OR LOWER(a.description) LIKE '%mutation%');

# 4. Step 2: Extract Activities
%sql DROP TABLE IF EXISTS ar_curated_data CASCADE;
%sql \
    SELECT act.molregno, act.pchembl_value, act.standard_value, \
           tmp.assay_chembl_id, tmp.year \
    INTO TEMPORARY TABLE ar_curated_data \
    FROM activities act \
    JOIN ar_raw_assays tmp ON act.assay_id = tmp.assay_id \
    WHERE act.standard_type = :measurement_type \
    AND act.standard_units = 'nM' \
    AND act.standard_relation = '=' \
    AND act.pchembl_value IS NOT NULL;

# 5. Step 3: Remove records with pChEMBL discrepancies >= 3.0
# We find molregnos where the difference between any two tests is too high
%sql \
    DELETE FROM ar_curated_data \
    WHERE molregno IN ( \
        SELECT t1.molregno \
        FROM ar_curated_data t1 \
        JOIN ar_curated_data t2 ON t1.molregno = t2.molregno \
        WHERE ABS(t1.pchembl_value - t2.pchembl_value) >= 3.0 \
    );

print("Discrepancy removal complete.")
```

```python
# 1. Configuration
%config SqlMagic.named_parameters="enabled"

# 2. Parameters
target_id = 'CHEMBL1871'
measurement = 'IC50'

print(f"Executing Maximal Curation for {target_id}...")

# ---------------------------------------------------------
# STEP 1: Gather Assays with Confidence 9 and Document Links
# ---------------------------------------------------------
%sql DROP TABLE IF EXISTS ar_assays_filtered CASCADE;
%sql \
    SELECT a.assay_id, a.doc_id, a.chembl_id as assay_chembl_id, \
           count(distinct act.molregno) as compound_count \
    INTO TEMPORARY TABLE ar_assays_filtered \
    FROM assays a \
    JOIN target_dictionary t ON a.tid = t.tid \
    JOIN activities act ON a.assay_id = act.assay_id \
    WHERE t.chembl_id = :target_id \
      AND t.target_type = 'SINGLE PROTEIN' \
      AND a.confidence_score = 9 \
      AND a.doc_id IS NOT NULL \
      AND a.assay_type = 'B' \
      AND a.variant_id IS NULL \
      AND NOT (LOWER(a.description) LIKE '%mutant%' OR LOWER(a.description) LIKE '%mutation%') \
    GROUP BY a.assay_id, a.doc_id, a.chembl_id;

# ---------------------------------------------------------
# STEP 2: Retain only the largest Assay per Document
# ---------------------------------------------------------
%sql DROP TABLE IF EXISTS ar_assays_final CASCADE;
%sql \
    SELECT assay_id, assay_chembl_id \
    INTO TEMPORARY TABLE ar_assays_final \
    FROM ( \
        SELECT assay_id, assay_chembl_id, \
               ROW_NUMBER() OVER (PARTITION BY doc_id ORDER BY compound_count DESC) as rank \
        FROM ar_assays_filtered \
    ) ranked_assays \
    WHERE rank = 1;

# ---------------------------------------------------------
# STEP 3: Extract Activities and Apply Discrepancy Filter (3.0 units)
# ---------------------------------------------------------
%sql DROP TABLE IF EXISTS ar_curated_data CASCADE;
%sql \
    SELECT act.molregno, act.pchembl_value, f.assay_chembl_id \
    INTO TEMPORARY TABLE ar_curated_data \
    FROM activities act \
    JOIN ar_assays_final f ON act.assay_id = f.assay_id \
    WHERE act.standard_type = :measurement \
      AND act.standard_units = 'nM' \
      AND act.standard_relation = '=' \
      AND act.pchembl_value IS NOT NULL;

# ---------------------------------------------------------
# STEP 4: Remove compounds with discrepancy >= 3.0 across assays
# ---------------------------------------------------------
%sql \
    DELETE FROM ar_curated_data \
    WHERE molregno IN ( \
        SELECT t1.molregno \
        FROM ar_curated_data t1 \
        JOIN ar_curated_data t2 ON t1.molregno = t2.molregno \
        WHERE ABS(t1.pchembl_value - t2.pchembl_value) >= 3.0 \
    );

print("Maximal curation complete.")
```
